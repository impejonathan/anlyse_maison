```python
# Obtenir les codes postaux uniques
zipcodes_uniques = df['zipcode'].unique()

# Boucle sur chaque code postal unique et ajoute une colonne correspondante √† df
for zipcode in zipcodes_uniques:
    # Cr√©er une nouvelle colonne pour le code postal
    df[zipcode] = pd.get_dummies(df['zipcode'].apply(lambda x: 1 if x==zipcode else 0))[1]
```
